```python
''' BEHAVIOR '''
# load behavioral data
for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        folder = base + m + '/'
        data_file = m + '_' + s + '_data.mat'
        behavior = get_data.loadData(folder + data_file)
        d['behavior'] = behavior
```

```python
# get reward and lick data
for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        behave_data = d['behavior']
        d['reward_data'] = {}
        
        # get behavioral params
        posx = behave_data['posx']
        post = behave_data['post']
        trial = get_data.trial_idx(posx)
        dt = np.unique(np.round(np.diff(post),4))
        speed = get_data.getSpeed(posx, dt)
        
        # save A unfiltered
        A = np.column_stack((posx, speed, trial, post))
        d['A_unfiltered'] = A
        
        # get reward params
        lickt = behave_data['lickt']
        reward = behave_data['reward']
        reward_centers = np.array(reward['centers'])
        rewardt = reward['times']
        auto_reward = reward['auto']
        misses = reward['miss'].astype(bool)
        reward_trials = reward['trials']
        
        # find estimated reward times for misses
        if np.sum(misses) > 0:
            missed_centers = reward_centers[misses]
            missed_trials = reward_trials[misses]
            missed_trials = missed_trials[missed_trials%100 > 0]
            for t in np.unique(missed_trials):
                trial_pos = posx[trial == t]
                trial_t = post[trial == t]
                trial_centers = missed_centers[missed_trials == t]

                for c in trial_centers:
                    idx = np.argmin(np.abs(trial_pos - c))
                    rewardt[(reward_trials == t) & (reward_centers == c)] = trial_t[idx]

        # get lick and reward indices
        dt = 0.02
        lick_ct = get_data.spiketrain(post, dt, lickt, index=False)
        reward_idx = get_data.spiketrain(post, dt, rewardt, index=True)
        d['reward_data']['lick_ct'] = lick_ct
        d['reward_data']['reward_idx'] = reward_idx
```

```python
''' in- vs. out-zone running speed '''
speed_reward = []
speed_out = []
for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
#         if 'block_ID' in d.keys():
#             continue
        behave_data = d['behavior']
        
        # get behavioral data
        A = d['A_unfiltered']
        posx = A[:, 0]
        trial = A[:, 2]
        speed = A[:, 1]
        if np.min(trial) > 0:
            trial = trial-1
        reward = behave_data['reward']
        reward_centers = np.array(reward['centers'])
        reward_trials = reward['trials']
        
        # get indices for reward zone edges
        start_idx = []
        end_idx = []
        for t, c in zip(reward_trials, reward_centers):
            zone_start = c - 25
            zone_end = c + 25
            trial_pos = posx.copy()
            trial_pos[trial != t] = 1000
            start_idx.append(np.argmin(np.abs(trial_pos - zone_start)))
            end_idx.append(np.argmin(np.abs(trial_pos - zone_end)))
        start_idx = np.asarray(start_idx)
        end_idx = np.asarray(end_idx)
        
        # get running speed within vs. outside reward zone
        in_zone_speed = np.asarray([])
        in_idx = np.asarray([])
        for i in range(start_idx.shape[0]):
            s = start_idx[i]
            e = end_idx[i] + 1
            in_zone_speed = np.append(in_zone_speed, speed[s:e])
            in_idx = np.append(in_idx, np.arange(s, e))
        out_idx = np.setdiff1d(np.arange(speed.shape[0]), in_idx)
        out_zone_speed = speed[out_idx]
        
        # save speeds
        speed_reward.append(in_zone_speed)
        speed_out.append(out_zone_speed)
```

```python
''' in- vs. out-zone running speed '''
for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        if 'block_ID' in d.keys():
            behave_data = d['behavior']

            # get behavioral data
            A = d['A_unfiltered']
            posx = A[:, 0]
            trial = A[:, 2]
            speed = A[:, 1]
            if np.min(trial) > 0:
                trial = trial-1
            reward = behave_data['reward']
            reward_centers = np.array(reward['centers'])
            reward_trials = reward['trials']
            
            # get block idx by trial and by obs
            cr_idx = d['block_ID']
            cr_idx_obs = np.zeros_like(speed)
            for t in np.unique(trial[cr_idx]):
                cr_idx_obs[trial==t] = 1
            cr_idx_obs = cr_idx_obs.astype(bool)

            # get indices for reward zone edges
            start_idx_cp = np.asarray([])
            end_idx_cp = np.asarray([])
            start_idx_cr = np.asarray([])
            end_idx_cr = np.asarray([])
            for t, c in zip(reward_trials, reward_centers):
                zone_start = c - 25
                zone_end = c + 25
                trial_pos = posx.copy()
                trial_pos[trial != t] = 1000
                if cr_idx[t]:
                    start_idx_cr = np.append(start_idx_cr, 
                                             np.argmin(np.abs(trial_pos - zone_start)))
                    end_idx_cr = np.append(end_idx_cr, 
                                           np.argmin(np.abs(trial_pos - zone_end)))
                else:
                    start_idx_cp = np.append(start_idx_cp, 
                                             np.argmin(np.abs(trial_pos - zone_start)))
                    end_idx_cp = np.append(end_idx_cp, 
                                           np.argmin(np.abs(trial_pos - zone_end)))
            start_idx_cp = start_idx_cp.astype(int)
            start_idx_cr = start_idx_cr.astype(int)
            end_idx_cp = end_idx_cp.astype(int)
            end_idx_cr = end_idx_cr.astype(int)
            
            # get running speed within vs. outside reward zone - split by cr/cp
            all_obs_idx = np.arange(speed.shape[0])
            
            #CUE POOR
            in_zone_speed_cp = np.asarray([])
            in_idx_cp = np.asarray([])
            for i in range(start_idx_cp.shape[0]):
                s = start_idx_cp[i]
                e = end_idx_cp[i] + 1
                in_zone_speed_cp = np.append(in_zone_speed_cp, speed[s:e])
                in_idx_cp = np.append(in_idx_cp, np.arange(s, e))
            out_idx_cp = np.setdiff1d(all_obs_idx[~cr_idx_obs], in_idx)
            out_zone_speed_cp = speed[out_idx_cp]
            speed_reward.append(in_zone_speed_cp)
            speed_out.append(out_zone_speed_cp)
            
            # CUE RICH
            in_zone_speed_cr = np.asarray([])
            in_idx_cr = np.asarray([])
            for i in range(start_idx_cr.shape[0]):
                s = start_idx_cr[i]
                e = end_idx_cr[i] + 1
                in_zone_speed_cr = np.append(in_zone_speed_cr, speed[s:e])
                in_idx_cr = np.append(in_idx_cr, np.arange(s, e))
            out_idx_cr = np.setdiff1d(all_obs_idx[cr_idx_obs], in_idx)
            out_zone_speed_cr = speed[out_idx_cr]
            speed_reward.append(in_zone_speed_cr)
            speed_out.append(out_zone_speed_cr)
```

```python
from scipy import stats

avg_speed_in = []
sem_speed_in = []
avg_speed_out = []
sem_speed_out = []
for speed_i, speed_o in zip(speed_reward, speed_out):
    avg_in = np.mean(speed_i)
    avg_speed_in.append(avg_in)
    sem_in = stats.sem(speed_i)
    sem_speed_in.append(sem_in)
    
    avg_out = np.mean(speed_o)
    avg_speed_out.append(avg_out)
    sem_out = stats.sem(speed_o)
    sem_speed_out.append(sem_out)  
    
avg_speed_in = np.asarray(avg_speed_in)
sem_speed_in = np.asarray(sem_speed_in)
avg_speed_out = np.asarray(avg_speed_out)
sem_speed_out = np.asarray(sem_speed_out)
```

```python
''' Wilcoxin signed rank in- vs. out-zone speed'''
diff_speed = avg_speed_out - avg_speed_in

print('mean, sem')
print('diff in speeds, out - in: ' + str(np.round(np.mean(diff_speed), 2)) + ', ' + str(np.round(stats.sem(diff_speed), 2)))

stat, p_diff = stats.wilcoxon(diff_speed)
print('\np = ' + str(p_diff))
```

```python
''' lick N in- vs. out-zone '''
lick_reward = []
lick_out = []

for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        behave_data = d['behavior']
        
        # get behavioral data
        A = d['A_unfiltered']
        posx = A[:, 0]
        trial = A[:, 2]
        if np.min(trial) > 0:
            trial = trial-1
        reward = behave_data['reward']
        reward_centers = np.array(reward['centers'])
        reward_trials = reward['trials']
        lick_ct = d['reward_data']['lick_ct']
        
        # get indices for reward zone edges
        start_idx = []
        end_idx = []
        for t, c in zip(reward_trials, reward_centers):
            zone_start = c - 25
            zone_end = c + 25
            trial_pos = posx.copy()
            trial_pos[trial != t] = 1000
            start_idx.append(np.argmin(np.abs(trial_pos - zone_start)))
            end_idx.append(np.argmin(np.abs(trial_pos - zone_end)))
        start_idx = np.asarray(start_idx)
        end_idx = np.asarray(end_idx)
        
        # get number of licks per reward zone
        in_zone_licks = np.zeros_like(start_idx)
        for i in range(start_idx.shape[0]):
            s = start_idx[i]
            e = end_idx[i] + 1
            in_zone_licks[i] = np.sum(lick_ct[s:e])

        # get number of licks per trial
        total_licks = np.zeros_like(np.unique(trial))
        for i, t in enumerate(np.unique(trial)):
            total_licks[i] = np.sum(lick_ct[trial==t])
            
        # get N in/out zone licks by reward trial
        in_zone = np.asarray([])
        out_zone = np.asarray([])
        for i, t in enumerate(np.unique(trial)):
            if t in reward_trials:
                total = total_licks[i]
                if total == 0:
                    in_zone = np.append(in_zone, 0)
                    out_zone = np.append(out_zone, 0)                    
                    continue
                licks_in = np.sum(in_zone_licks[np.where(reward_trials==t)[0]])
                in_zone = np.append(in_zone, licks_in)
                out_zone = np.append(out_zone, total - licks_in)
        
        # save licks
        lick_reward.append(in_zone)
        lick_out.append(out_zone)
```

```python
''' lick N in- vs. out-zone '''
for m, session in zip(mice, sessions):
    for s in session:
        d = data[m][s]
        if 'block_ID' in d.keys():
            behave_data = d['behavior']
            behave_data = d['behavior']

            # get behavioral data
            A = d['A_unfiltered']
            posx = A[:, 0]
            trial = A[:, 2]
            if np.min(trial) > 0:
                trial = trial-1
            reward = behave_data['reward']
            reward_centers = np.array(reward['centers'])
            reward_trials = reward['trials']
            lick_ct = d['reward_data']['lick_ct']

            # get block idx by trial and by obs
            cr_idx = d['block_ID'].astype(bool)
            
            # get indices for reward zone edges
            start_idx = []
            end_idx = []
            for t, c in zip(reward_trials, reward_centers):
                zone_start = c - 25
                zone_end = c + 25
                trial_pos = posx.copy()
                trial_pos[trial != t] = 1000
                start_idx.append(np.argmin(np.abs(trial_pos - zone_start)))
                end_idx.append(np.argmin(np.abs(trial_pos - zone_end)))
            start_idx = np.asarray(start_idx)
            end_idx = np.asarray(end_idx)

            # get number of licks per reward zone
            in_zone_licks = np.zeros_like(start_idx)
            for i in range(start_idx.shape[0]):
                s_idx = start_idx[i]
                e = end_idx[i] + 1
                in_zone_licks[i] = np.sum(lick_ct[s_idx:e])

            # get number of licks per trial
            unique_trials = np.unique(trial)
            total_licks = np.zeros_like(unique_trials)
            for i, t in enumerate(unique_trials):
                total_licks[i] = np.sum(lick_ct[trial==t])

            # get N in/out zone licks by reward trial
            # CUE POOR
            in_zone = np.asarray([])
            out_zone = np.asarray([])
            for i, t in enumerate(unique_trials[~cr_idx]):
                if t in reward_trials:
                    total = total_licks[unique_trials==t]
                    if total == 0:
                        in_zone = np.append(in_zone, 0)
                        out_zone = np.append(out_zone, 0)                    
                        continue
                    licks_in = np.sum(in_zone_licks[np.where(reward_trials==t)[0]])
                    in_zone = np.append(in_zone, licks_in)
                    out_zone = np.append(out_zone, total - licks_in)
                    if licks_in > total:
                        print('{}_{}'.format(m, s))
                        print('trial {}'.format(t))
                        print('in zone licks = {}'.format(licks_in))
                        print('total licks = {}'.format(total))
    
            # save licks
            lick_reward.append(in_zone)
            lick_out.append(out_zone)
            
            # CUE RICH
            in_zone = np.asarray([])
            out_zone = np.asarray([])
            for i, t in enumerate(unique_trials[cr_idx]):
                if t in reward_trials:
                    total = total_licks[unique_trials==t]
                    if total == 0:
                        in_zone = np.append(in_zone, 0)
                        out_zone = np.append(out_zone, 0)                    
                        continue
                    licks_in = np.sum(in_zone_licks[np.where(reward_trials==t)[0]])
                    in_zone = np.append(in_zone, licks_in)
                    out_zone = np.append(out_zone, total - licks_in)

            # save licks
            lick_reward.append(in_zone)
            lick_out.append(out_zone)
```

```python
avg_lick_in = []
sem_lick_in = []
avg_lick_out = []
sem_lick_out = []
diff_licks = np.asarray([])
for lick_i, lick_o in zip(lick_reward, lick_out):
    # normalize
    total_licks = lick_i + lick_o
    norm_lick_i = lick_i / total_licks
    norm_lick_o = lick_o / total_licks
    norm_lick_i[np.isnan(norm_lick_i)] = 0
    norm_lick_o[np.isnan(norm_lick_o)] = 0
    
    avg_lick_in.append(np.mean(norm_lick_i))
    sem_lick_in.append(stats.sem(norm_lick_i))    
    avg_lick_out.append(np.mean(norm_lick_o))
    sem_lick_out.append(stats.sem(norm_lick_o))
    
    diff_licks = np.append(diff_licks, lick_i - lick_o)
    
avg_lick_in = np.asarray(avg_lick_in)
sem_lick_in = np.asarray(sem_lick_in)
avg_lick_out = np.asarray(avg_lick_out)
sem_lick_out = np.asarray(sem_lick_out)
```

```python
''' Wilcoxin signed rank in- vs. out-zone N licks'''
print('mean, sem')
print('diff in lick n, out - in: ' + str(np.round(np.mean(diff_licks), 2)) + ', ' + str(np.round(stats.sem(diff_licks), 2)))
print('n reward trials = ' + str(diff_licks.shape[0]))

stat, p_diff = stats.wilcoxon(diff_licks)
print('\np = ' + str(p_diff))
```

```python
'''
explore local versus distant similarity:
- average score for k = 1:5 (nearby, spatial stability)
- np.diff(flatten upper triangle) histogram--bimodal? (large transitions)
'''
max_k = 5
avg_k = np.zeros(N_sessions)
k_25 = np.zeros(N_sessions)
k_75 = np.zeros(N_sessions)
i = -1
for m, session in zip(mice, sessions):
    for s in session:
        i += 1
        d = data[m][s]
        sim = d['similarity']

        sum_k = np.zeros(sim.shape[0])
        for k in np.arange(-max_k, max_k+1, 1):
            if k == 0:
                continue
            elif k < 0:
                sum_k[np.abs(k):] = sum_k[np.abs(k):] + np.diagonal(sim, offset=k)
            else:
                sum_k[:-k] = sum_k[:-k] + np.diagonal(sim, offset=k)
        local_stability = sum_k[max_k:-max_k]/(2*max_k)
        d['local_stability'] = local_stability
        
        avg_k[i] = np.mean(local_stability)
        k_25[i] = np.percentile(local_stability, 25)
        k_75[i] = np.percentile(local_stability, 75)
```

```python
# calculate local stability for cue rich vs. poor blocks
max_k = 5

avg_k_CR = np.zeros(N_double)
k_25_CR = np.zeros(N_double)
k_75_CR = np.zeros(N_double)
avg_k_CP = np.zeros(N_double)
k_25_CP = np.zeros(N_double)
k_75_CP = np.zeros(N_double)
i = -1
for m, session in zip(mice[-5:], sessions[-5:]):
    for s in session:
        i += 1
        d = data[m][s]
        sim = d['similarity']
        block_ID = d['block_ID'].astype(bool)
        sim_CR = sim[block_ID, :]
        sim_CR = sim_CR[:, block_ID]
        sim_CP = sim[~block_ID, :]
        sim_CP = sim_CP[:, ~block_ID]

        #cue rich
        sum_k = np.zeros(sim_CR.shape[0])
        for k in np.arange(-max_k, max_k+1, 1):
            if k == 0:
                continue
            elif k < 0:
                sum_k[np.abs(k):] = sum_k[np.abs(k):] + np.diagonal(sim_CR, offset=k)
            else:
                sum_k[:-k] = sum_k[:-k] + np.diagonal(sim_CR, offset=k)
        local_stability = sum_k[max_k:-max_k]/(2*max_k)
        avg_k_CR[i] = np.mean(local_stability)
        k_25_CR[i] = np.percentile(local_stability, 25)
        k_75_CR[i] = np.percentile(local_stability, 75)
        
        #cue poor
        sum_k = np.zeros(sim_CP.shape[0])
        for k in np.arange(-max_k, max_k+1, 1):
            if k == 0:
                continue
            elif k < 0:
                sum_k[np.abs(k):] = sum_k[np.abs(k):] + np.diagonal(sim_CP, offset=k)
            else:
                sum_k[:-k] = sum_k[:-k] + np.diagonal(sim_CP, offset=k)
        local_stability = sum_k[max_k:-max_k]/(2*max_k)
        avg_k_CP[i] = np.mean(local_stability)
        k_25_CP[i] = np.percentile(local_stability, 25)
        k_75_CP[i] = np.percentile(local_stability, 75)
```
